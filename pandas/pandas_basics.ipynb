```python
# DataFrame:

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Alice"],
    "Age": [25, 30, 35, np.nan, 29, 25],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "HR"],
    "Salary": [50000, 60000, 70000, 62000, np.nan, 50000]
}
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

```python
# loc and iloc
df.iloc[0:3, 0:4:2] #rows, columns
```

```python
df.loc[1:4,['Name','Salary']]
```

```python
df['Age']
```

```python
df[['Age','Salary']]
```

```python
df.drop("Age",axis=1)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Broadcasting
df['Salary'] = df["Salary"] + 5000
df['Salary']
```

```python
# Renaming columns
df.rename(columns={'Salary':'Pay'},inplace=True)
df
```

```python
df ['Department'].unique()
```

```python
df["Department"].value_counts()
```

```python
# Adding new columns
df["Promoted Salary"] = df["Pay"] + df["Pay"] * 10 / 100
df
```

```python
# Data Cleaning
df.dropna(how = "any") #Drop any row that has any null value
```

```python
df.dropna(how="all") #Drop only if all the values in a row are null
```

```python
#Dealing with missing values by filling them up
df["Age"].fillna(df['Age'].mean())
```

```python
df["Pay"].fillna(df['Pay'].median())
```

```python
# Using forward and backward fill
df["Age"].fillna(method="ffill")
```

```python
df['Age'].fillna(method='bfill')
```

```python
df
```

```python
#Replacing values
df["Name"].replace("Alice","Spparke")
```

```python
df["Name"] = df["Name"].replace("Alice","Spparke") # Makes changes to original dataset
df
```

```python
# Duplicates
df_duplicate = df[df.duplicated(keep = "last")] #Default parameter is keep="first"
df_duplicate
```

```python
df = df.drop_duplicates()
df
```

```python
# Invalid values
# Lambda -> python
df["Promoted Salary"] = df["Promoted Salary"].apply(lambda x: x-5000 if x>82000 else x)
df
```

```python
name="alex_spies"
df[["first_name", "last_name"]] = df["name"].str.split("_") #Will work if there is a column named 'name', and contains value in the form
                                                            # 'xxx_yyy' which will further produce xxx as 'first_name' and yyy as 'last_name'
```

```python
#Apply and lambda
df["Age"] = df["Age"].apply(lambda x: x*2)
df
```

```python
def original_age(x):
  return x/2
df["Age"] = df["Age"].apply(original_age)
df
```

```python
# Joins and Merges

department_info = {
    "Dept": ["HR", "IT", "Finance"],
    "Location": ["New York", "San Francisco", "Chicago"],
    "Manager": ["Laura", "Steve", "Nina"]
}

df2 = pd.DataFrame(department_info)
df2
```

```python
pd.concat([df,df2])
```

```python
pd.concat([df,df2],axis=1)
```

```python
pd.merge(df, df2, left_on='Department', right_on='Dept')
```
